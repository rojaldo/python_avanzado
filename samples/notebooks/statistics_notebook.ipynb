# Conceptos Estadísticos
1. **Media**: El promedio de un conjunto de datos, calculado sumando todos los valores y dividiendo entre la cantidad de valores.

2. **Mediana**: El valor central de un conjunto de datos cuando está ordenado de menor a mayor. Si el conjunto tiene un número par de valores, la mediana es el promedio de los dos valores centrales.

3. **Moda**: El o los valores que aparecen con mayor frecuencia en un conjunto de datos. Un conjunto puede tener una moda, más de una moda o ninguna moda.

4. **Rango**: La diferencia entre el valor máximo y el valor mínimo en un conjunto de datos.

5. **Cuartiles**:
    - **Q1 (Primer Cuartil)**: El valor por debajo del cual se encuentra el 25% de los datos.
    - **Q2 (Segundo Cuartil)**: La mediana del conjunto de datos.
    - **Q3 (Tercer Cuartil)**: El valor por debajo del cual se encuentra el 75% de los datos.
    - **Rango Intercuartílico (IQR)**: La diferencia entre Q3 y Q1, que representa la dispersión del 50% central de los datos.

6. **Varianza**: Una medida de cuánto varían los valores de un conjunto de datos respecto a la media. Se calcula como el promedio de las diferencias al cuadrado respecto a la media.

7. **Desviación Estándar (Std)**: La raíz cuadrada de la varianza. Proporciona una medida de la dispersión de los datos en las mismas unidades que los datos.

8. **Asimetría**: Una medida de la falta de simetría en la distribución de los valores de un conjunto de datos. Una asimetría positiva indica una cola más larga a la derecha, mientras que una asimetría negativa indica una cola más larga a la izquierda.

9. **Curtosis**: Una medida de la "apuntamiento" de la distribución. Una curtosis alta indica colas pesadas, mientras que una curtosis baja indica colas ligeras.

10. **Correlación**: Una medida de la relación entre dos variables, que varía de -1 (correlación negativa perfecta) a 1 (correlación positiva perfecta).

11. **Puntaje Z (Z-Score)**: El número de desviaciones estándar que un punto de datos está por encima o por debajo de la media. Se utiliza para estandarizar datos.

12. **Percentiles**: Valores por debajo de los cuales se encuentra un cierto porcentaje de los datos. Por ejemplo, el percentil 90 es el valor por debajo del cual se encuentra el 90% de los datos.

13. **Valores Atípicos (Outliers)**: Puntos de datos que son significativamente diferentes del resto del conjunto. Pueden identificarse utilizando métodos como la regla del IQR o los puntajes Z.

Estos conceptos son fundamentales para entender y analizar datos en estadística.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
```

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

# Distribución Normal
normal_data = np.random.normal(0, 1, 1000)
axs[0, 0].hist(normal_data, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
axs[0, 0].set_title("Distribución Normal")
axs[0, 0].set_xlabel("Valor")
axs[0, 0].set_ylabel("Frecuencia")

# Distribución Uniforme
uniform_data = np.random.uniform(0, 10, 1000)
axs[0, 1].hist(uniform_data, bins=30, color='lightgreen', edgecolor='black', alpha=0.7)
axs[0, 1].set_title("Distribución Uniforme")
axs[0, 1].set_xlabel("Valor")
axs[0, 1].set_ylabel("Frecuencia")

# Distribución Exponencial
exponential_data = np.random.exponential(scale=2, size=1000)
axs[1, 0].hist(exponential_data, bins=30, color='orange', edgecolor='black', alpha=0.7)
axs[1, 0].set_title("Distribución Exponencial")
axs[1, 0].set_xlabel("Valor")
axs[1, 0].set_ylabel("Frecuencia")

# Distribución Poisson
poisson_data = np.random.poisson(3, 1000)
axs[1, 1].hist(poisson_data, bins=30, color='lavender', edgecolor='black', alpha=0.7)
axs[1, 1].set_title("Distribución Poisson")
axs[1, 1].set_xlabel("Valor")
axs[1, 1].set_ylabel("Frecuencia")

plt.tight_layout()
plt.show()
```

```python
# Generamos una distribución normal con media 0 y desviación estándar 1
data = np.random.normal(loc=0, scale=1, size=1000)

# Calculamos la media de la distribución
media = np.mean(data)

# Creamos un histograma de la distribución
plt.hist(data, bins=30, alpha=0.7, color='blue', edgecolor='black')

# Dibujamos una línea vertical en la posición de la media
plt.axvline(media, color='red', linestyle='dashed', linewidth=2, label=f'Media = {media:.2f}')

# Agregamos etiquetas y título
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title('Distribución Normal y su Media')
plt.legend()

plt.show()
```

```python
# Generar una distribución exponencial (sesgada a la derecha)
data_skewed = np.random.exponential(scale=2, size=1000)

# Calcular la media y la mediana de la distribución sesgada
media_skewed = np.mean(data_skewed)
mediana_skewed = np.median(data_skewed)

# Crear un histograma de la distribución
plt.hist(data_skewed, bins=30, alpha=0.7, color='orange', edgecolor='black')

# Dibujar líneas verticales para la media y la mediana
plt.axvline(media_skewed, color='red', linestyle='dashed', linewidth=2, label=f'Media = {media_skewed:.2f}')
plt.axvline(mediana_skewed, color='green', linestyle='dashed', linewidth=2, label=f'Mediana = {mediana_skewed:.2f}')

# Configurar etiquetas y título del gráfico
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.title('Distribución Normal: Media vs Mediana')
plt.legend()

plt.show()
```

```python
# Calcular los cuartiles para la distribución sesgada
q1_skewed, mediana_skewed_calculated, q3_skewed = np.percentile(data_skewed, [25, 50, 75])

# Crear el histograma de data_skewed
plt.hist(data_skewed, bins=30, alpha=0.7, color='orange', edgecolor='black')

# Dibujar líneas verticales para cada cuartil
plt.axvline(q1_skewed, color='blue', linestyle='dashed', linewidth=2, label=f"Q1 = {q1_skewed:.2f}")
plt.axvline(mediana_skewed_calculated, color='green', linestyle='dashed', linewidth=2, label=f"Mediana = {mediana_skewed_calculated:.2f}")
plt.axvline(q3_skewed, color='red', linestyle='dashed', linewidth=2, label=f"Q3 = {q3_skewed:.2f}")

# Rellenar la región entre Q1 y Q3 (el IQR)
ylim = plt.gca().get_ylim()
plt.fill_betweenx(np.linspace(ylim[0], ylim[1], 100), q1_skewed, q3_skewed, color='gray', alpha=0.2, label="IQR")

plt.xlabel("Valor")
plt.ylabel("Frecuencia")
plt.title("Histograma de 'data_skewed' con Cuartiles")
plt.legend()
plt.show()
```

```python

# Generamos datos correlacionados
np.random.seed(42)  # para reproducibilidad
x_corr = np.random.normal(0, 1, 1000)
y_corr = 0.8 * x_corr + np.random.normal(0, 0.5, 1000)

# Calculamos el coeficiente de correlación de Pearson
corr_coef_corr, _ = pearsonr(x_corr, y_corr)

# Creamos una figura con 3 subplots: histogramas de x, de y y diagrama de dispersión
fig, axs = plt.subplots(1, 3, figsize=(15, 4))

axs[0].hist(x_corr, bins=30, color='skyblue', edgecolor='black')
axs[0].set_title("Distribución de X Correlacionada")

axs[1].hist(y_corr, bins=30, color='lightgreen', edgecolor='black')
axs[1].set_title("Distribución de Y Correlacionada")

axs[2].scatter(x_corr, y_corr, alpha=0.5)
axs[2].set_xlabel("X Correlacionada")
axs[2].set_ylabel("Y Correlacionada")
axs[2].set_title(f"Diagrama de dispersión\nCoef. Correlación: {corr_coef_corr:.2f}")

plt.tight_layout()
plt.show()
```

```python
# Ejemplo: Visualización de 3 casos de Puntaje Z (Z-Score)
# Caso 1: Valor típico (cercano a 0)
# Caso 2: Valor alto (posible outlier positivo)
# Caso 3: Valor bajo (posible outlier negativo)

import matplotlib.pyplot as plt

# Definir 3 valores de Z-score que ejemplifican los casos
z_normal = 0.0     # valor típico
z_alto   = 2.0     # valor alto
z_bajo   = -2.0    # valor bajo

# Creamos una figura con 3 subplots
fig, axs = plt.subplots(1, 3, figsize=(15, 4), sharey=True)

# Caso 1: Valor típico (Z cercano a 0)
axs[0].scatter(z_normal, 0, color='blue', s=150)
axs[0].axvline(z_normal, color='black', linestyle='dotted')
axs[0].set_title("Caso 1: Valor Típico\n(z ≈ 0)")
axs[0].set_xlim(-3, 3)
axs[0].set_ylim(-1, 1)
axs[0].set_xlabel("Puntaje Z")
axs[0].text(z_normal, 0.3, "Normal", color='blue',
            horizontalalignment='center')

# Caso 2: Valor Alto (posible outlier positivo)
axs[1].scatter(z_alto, 0, color='red', s=150)
axs[1].axvline(z_alto, color='black', linestyle='dotted')
axs[1].set_title("Caso 2: Valor Alto\n(z = 2)")
axs[1].set_xlim(-3, 3)
axs[1].set_ylim(-1, 1)
axs[1].set_xlabel("Puntaje Z")
axs[1].text(z_alto, 0.3, "Potencial outlier", color='red',
            horizontalalignment='center')

# Caso 3: Valor Bajo (posible outlier negativo)
axs[2].scatter(z_bajo, 0, color='green', s=150)
axs[2].axvline(z_bajo, color='black', linestyle='dotted')
axs[2].set_title("Caso 3: Valor Bajo\n(z = -2)")
axs[2].set_xlim(-3, 3)
axs[2].set_ylim(-1, 1)
axs[2].set_xlabel("Puntaje Z")
axs[2].text(z_bajo, 0.3, "Potencial outlier", color='green',
            horizontalalignment='center')

fig.suptitle("Ejemplos de Puntaje Z (Z-Score)")
plt.tight_layout(rect=[0, 0, 1, 0.92])
plt.show()
```
